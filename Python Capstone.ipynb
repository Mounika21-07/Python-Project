```python
pd.read_csv('project_df.csv') #using the project dataframe which was created already

project_df['Cost'] = pd.to_numeric(project_df['Cost'], errors='coerce')

# Compute the average of non-missing values to replace it with the missing values

total = 0
count = 0
'''Using for loop to calculate the total and count to use it to find the average
'''
for value in project_df['Cost']:
   if not pd.isna(value):  # Check for non-NaN values to find average
        total += value
        count += 1

average = total / count if count > 0 else 0  # to avoid division by zero

# Replace missing values with the average using a FOR loop

for i in range(len(project_df['Cost'])):
    if pd.isna(project_df.loc[i, 'Cost']):  # Checking for NaN value
        project_df.loc[i, 'Cost'] = average  # Replacing NaN with the average value
        

project_df1 = project_df.copy() #creating a copy to showcase the updated dataframe

project_df1['Cost'] = project_df1['Cost'].astype(int) #changing the data type from float to int

# Display the updated DataFrame
project_df1
```

```python
Employee_df = pd.read_csv('Employee_df.csv')

# Step 1: Split the 'name' column into 'firstname' and 'lastname'

Employee_df[['firstname', 'lastname']] = Employee_df['Name'].str.split(' ', expand=True)

'''str.split is used to split the coloum name values into first name and last name with space as delimeter'''

# Step 2: Remove the original 'name' column using drop function

Employee_df.drop('Name', axis=1, inplace=True)

# Display the updated DataFrame by creating a copy of dataframe

Employee_df1= Employee_df.copy()

Employee_df1
```

```python
'''Reading the three different dataframes which were created'''

project_df = pd.read_csv('project_df.csv')
seniority_level_df= pd.read_csv('seniority_level_df.csv')
Employee_df = pd.read_csv('Employee_df.csv')

'''Using merge function to join the project and employee dataframe with inner join type and merging it on the basis of ID column'''

Final1 = pd.merge(project_df1,Employee_df1,on = 'ID',how='inner')

Final = pd.merge(Final1,seniority_level_df, on = 'ID', how='inner') 

'''merging the final 1 and the third dataframe to get the output of all three dataframes'''

Final
```

```python

'''Definfing the function as bonus_calculation and taking row as parameter'''

def bonus_calculation(row):
    if row['Status'] == 'Finished':   
        return row['Cost']*0.05
    else:
        return 0

'''applying the if condition using comparison parameter to apply the bonus calculation on the finished projects'''

Final['Bonus'] = Final.apply(bonus_calculation,axis=1) 

'''applying the function bonus_calculation in the rows and creating the column bonus'''

Final
```

```python

'''creating the copy of the dataframe final to avoid reassignment of code'''

Final1 = Final.copy()

Final1.loc[Final1['Status'] == 'Failed', 'Designation Level'] += 1

'''Demote the designation of employees with project status as failed, using +=1 to increase the designation level as higher number means lower designation'''

Delete_rows = Final1[Final1['Designation Level'] >4].index

'''using drop function to delete the rows of employees having designation level greater than 4'''

Final1.drop(Delete_rows,inplace = True)
Final1
```

```python

'''creating the copy of the dataframe final to avoid reassignment of code'''

Final3 = Final.copy()

'''Add Mr. or Mrs. to the FirstName based on the Gender'''

Final3["firstname"] = Final3.apply(lambda row: "Mr. " + row["firstname"] if row["Gender"] == "M" else "Mrs. " + row["firstname"], axis=1)

'''applying conditions to add Mr. when gender is M,else Mrs.'''

'''Using lambda func to define a func inside a func'''

Final3.drop("Gender", axis=1, inplace=True) 

'''deleting the gender column after making the changes using drop function'''

Final3
```

```python
'''creating the copy of the dataframe final to avoid reassignment of code'''

Final4= Final.copy()

'''if the age is greater than 29, promoting the designation level(-1), apply is used to apply the function to the rows taken in the lambda '''

Final4['Designation Level'] = Final4.apply(lambda row: row['Designation Level'] - 1 
                                           if row['Age'] > 29 else row['Designation Level'], axis=1)


'''printing the dataframe'''

Final4 
```

```python
'''creating a new dataframe 'TotalProjCost'using Group by ID and 'First Name' in the final dataframe, summing the costs'''

TotalProjCost = Final.groupby(['ID', 'firstname'], as_index=False)['Cost'].sum()

'''Rename the 'Cost' column to 'Total Cost'''

TotalProjCost.rename(columns={'Cost': 'Total Cost'}, inplace=True)

TotalProjCost
```
